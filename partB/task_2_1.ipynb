## Dataset Choice and Justification

For this experiment I use a synthetic dataset generated using `sklearn.datasets.make_classification`.

The dataset contains **100 samples and 20 numerical features**, of which only **2 features are informative** while the remaining features introduce noise. This creates a relatively challenging learning scenario where the predictive signal is weak and spread across a higher-dimensional feature space.

This dataset is appropriate for reproducing the CW-SVM method because the algorithm is designed to incorporate **expert-labeled feature rankings** to guide the learning process when labeled data is limited. By using a dataset with few informative features and a relatively small training set, we simulate a situation where prior knowledge about feature importance can influence the learned classifier.

However, compared to the datasets used in the original paper (such as biomedical citation screening datasets and movie review sentiment datasets), this toy dataset is still much simpler. The original experiments operate on very high-dimensional bag-of-words text representations with thousands of sparse features, whereas this synthetic dataset contains only 20 dense numerical features. Therefore, the goal here is not to reproduce the exact performance reported in the paper but to demonstrate the principle of incorporating feature-level knowledge into the learning process.



```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

np.random.seed(42)

X, y = make_classification(
    n_samples=100,
    n_features=20,
    n_informative=2,
    n_redundant=2,
    n_classes=2,
    random_state=42
)

print("Dataset shape:", X.shape)
```

## Train-Test Split

The dataset is split into training and testing sets so that model performance can be evaluated on unseen data.  
A **large test split (60%)** is used so that the training set contains only **40 samples**, which simulates a **low-data scenario**. This setting is relevant because CW-SVM is designed to be particularly helpful when labeled training data is scarce and expert knowledge can provide additional guidance.


```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.6,
    random_state=42
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

## Preprocessing

Only minimal preprocessing is required for this dataset. The features generated by `make_classification` are already numerical and can be used directly by the Support Vector Machine.

No feature scaling is applied in this experiment. This is intentional because the reproduction simulates ranked feature constraints by manually adjusting the magnitude of selected features. Applying feature scaling would normalize these magnitudes and weaken the effect of the simulated ranking.

## Feature Representation

Each row of **X** represents one data instance and each column corresponds to a feature.

In later tasks, a subset of these features will be treated as **expert-labeled features**. Their relative importance will be manually adjusted to simulate the **ranked feature constraints** described in the CW-SVM paper. This allows us to demonstrate how incorporating feature-level knowledge can influence the learned classifier.
