# Task 1.2 — Key Assumptions

Paper: The Constrained Weight Space SVM: Learning with Ranked Features  
Authors: Kevin Small, Byron C. Wallace, Carla E. Brodley, Thomas A. Trikalinos  
Venue: ICML 2011

---

## Assumption 1

**Assumption:**  
The expert-provided ranking of features correctly reflects how strongly those features correlate with the class labels.

**Why the method needs it:**  
CW-SVM enforces constraints between feature weights based on the expert ranking. If the ranking is incorrect, the optimisation process will enforce the wrong ordering of weights, which can reduce the model’s classification performance.

**Violation scenario:**  
For example, in biomedical document screening an expert might mistakenly rank a weakly relevant term higher than a strongly relevant one. The constraint would then force the model to give the weaker feature a higher weight, harming the classifier.

**Paper reference:**  
Section 3.1.2 (Ranked Features), where constraints of the form wα − wβ ≥ ρ are applied between ranked features.

---

## Assumption 2

**Assumption:**  
Each feature dimension corresponds to a clearly identifiable feature such as a word in a bag-of-words representation.

**Why the method needs it:**  
CW-SVM applies constraints directly to individual components of the weight vector. This only works if each weight corresponds to a specific interpretable feature. If the representation used dense embeddings or latent features, these constraints would no longer correspond to meaningful concepts.

**Violation scenario:**  
If features were neural embeddings where each dimension represents a mixture of words, a constraint meant for a specific term would instead affect an arbitrary latent dimension.

**Paper reference:**  
Section 2 (Preliminaries), where the input vector is defined as x ∈ {0,1}^d, and Section 4.2 where bag-of-words features are used.

---

## Assumption 3

**Assumption:**  
A small set of labeled features is sufficient to guide the learning process.

**Why the method needs it:**  
The CW-SVM framework assumes that even a small number of constrained features can influence the rest of the weight vector during optimisation. This allows the model to benefit from expert knowledge without requiring a large amount of labeled training data.

**Violation scenario:**  
If the dataset’s predictive signal is spread across many features and the labeled ones are not representative, the constraints will not provide meaningful guidance and CW-SVM may not outperform a standard SVM.

**Paper reference:**  
Section 3 and Section 6 discuss how only a small number of labeled features are needed to influence the classifier.
